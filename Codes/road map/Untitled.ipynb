```python
import json

lats = []
lngs = []
with open('roads.geojson') as f:
    data = json.load(f)
index = 0
for feature in data['features']:
    for coordinate in feature['geometry']['coordinates']:
        #print (coordinate[0])
        lats.append(coordinate[0])
        lngs.append(coordinate[1])
    #print (feature['geometry']['coordinates'])
    #print (feature)
    if(index==3):
        break
    index+=1
```
